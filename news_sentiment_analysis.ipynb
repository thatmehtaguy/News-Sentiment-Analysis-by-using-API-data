```python
news = response.json()['articles']
```

```python
titles = []
content = []

for i in news:
    title = i['title']
    titles.append(title)
    cont = i['content']
    content.append(cont)
```

```python
import pandas as pd
articles = pd.DataFrame({'title':titles,'content':content})
```

```python
articles.head()
```

```python
articles['content'] = articles['content'].str.lower()
articles.head()
```

```python
articles['content'] = articles['content'].str.replace(r'[^\w\s]+', '')
articles.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
```

```python
articles['content'] = articles['content'].apply(lambda x: remove_stopwords(x))
articles.head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
import nltk
nltk.download('wordnet')
```

```python
import nltk
nltk.download('omw-1.4')
```

```python
articles['content'] = articles['content'].apply(lambda text: lemmatize_words(text))
articles.head()
```
